```python
import numpy as np
import re
import matplotlib.pyplot as plt
```

# Part 1

```python
input = np.loadtxt('input.txt', dtype=str)
```

```python
dir_reg = re.compile(r'(\w)(\d+)')

def parse_input(input):
    dirs = []
    for line in input:
        match = dir_reg.match(line)
        if match:
            dirs.append((match.group(1), int(match.group(2))))
    return dirs

parsed_input = parse_input(input)
```

```python
initial = 50
current = 50
```

```python
all_num = []
current = 50
count = 0
for input_line in parsed_input:
    
    if input_line[0] == 'R':
        # print("Moving to the Right by",input_line[1],"from", current)
        current = current + input_line[1]
        current = current % 100
        
    if input_line[0] == 'L':
        # print("Moving to the Left by",input_line[1],"from", current)
        current = current - input_line[1]
        current = current % 100
    
    if current == 0:
        count += 1
    all_num.append(current)
print(count)
```

```python
all_num = []
current = 50
count = 0 
pass_count = 0

for input_line in parsed_input:
    dist = input_line[1]

    if input_line[0] == 'R':
        pass_count += (current + dist - 1) // 100
        current = (current + dist) % 100

    elif input_line[0] == 'L':
        pass_count += (current - 1) // 100 - (current - dist) // 100
        current = (current - dist) % 100
    
    if current == 0:
        count += 1
        
    all_num.append(current)

print(count)
print(pass_count)
print(count + pass_count)
```
